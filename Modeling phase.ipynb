```python
#DADES FASE NO INICIAL
import pandas as pd
dfq3_new = pd.read_csv('qfasenoini19_net.csv') ##Q3 - Third semester data
dfinitial_new = pd.read_csv('qfaseini19_net.csv') #Initial phase data
dfpre_new = pd.read_csv("dpersnomespreins19esc_net.csv") #Preregistration data
df_times_initial = pd.read_csv("times_initial.csv") #Table with number of times a subject was taken
df_initial_average = pd.read_csv("average_initial.csv") #Table with average grades of the initial phase

#MERGE THE DATAFRAMES
inner = pd.merge(dfpre_new, dfinitial_new, on='CODI_EXPEDIENT', how='inner')
merged = pd.merge(inner, df_times_initial, on='CODI_EXPEDIENT', how='inner')
initial_merge = pd.merge(merged, df_initial_average, on='CODI_EXPEDIENT',how='inner')
total_merged = pd.merge(initial_merge, dfq3_new, on='CODI_EXPEDIENT')
total_merged.drop(columns=['Unnamed: 0','CODI_EXPEDIENT'], inplace=True)
total_merged.to_csv("merged.csv")
total_merged.head()
```

```python
# SPLITTING THE DATA AND FIRST EVALUATION
## This code is meant to treat the data for training through MLPClassifier and holdout
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

#Calling the output variables
y_phy = total_merged[['Physics']] 
y_mat = total_merged[['Materials']] 
y_em = total_merged[['Electromagnetism']] 
y_nm = total_merged[['Numerical methods']] 
y_de = total_merged[['Differential equations']] 
y_cs = total_merged[['Computer science II']] 

#Calling the input variables
X = total_merged.drop(['Electromagnetism','Numerical methods','Materials','Differential equations','Computer science II','Physics'],axis='columns') #everything else

#Splitting the data into training and testing sets through Holdout method
X_train_phy, X_test_phy, y_train_phy, y_test_phy = train_test_split(X, y_phy, test_size = 0.2, random_state = 42, stratify=y_phy)
X_train_mat, X_test_mat, y_train_mat, y_test_mat = train_test_split(X, y_mat, test_size = 0.2, random_state = 42, stratify=y_mat)
X_train_em, X_test_em, y_train_em, y_test_em = train_test_split(X, y_em, test_size = 0.2, random_state = 42, stratify=y_em)
X_train_nm, X_test_nm, y_train_nm, y_test_nm = train_test_split(X, y_nm, test_size = 0.2, random_state = 42, stratify=y_nm)
X_train_de, X_test_de, y_train_de, y_test_de = train_test_split(X, y_de, test_size = 0.2, random_state = 42, stratify=y_de)
X_train_cs, X_test_cs, y_train_cs, y_test_cs = train_test_split(X, y_cs, test_size = 0.2, random_state = 42, stratify=y_cs)
```

```python
#Saving the training and testing data for PHYSICS
X_train_phy.to_csv("X_train_phy.csv")
X_test_phy.to_csv("X_test_phy.csv")

#creating dataframes for the outputs (y):
df_y_train_phy = pd.DataFrame(y_train_phy,columns=['Physics'],index=X_train_phy.index)
df_y_test_phy = pd.DataFrame(y_test_phy,columns=['Physics'],index=X_test_phy.index)
df_y_train_phy.to_csv("y_train_phy.csv")
df_y_test_phy.to_csv("y_test_phy.csv")
```
